````python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

for layer in base_model.layers:
    layer.trainable = False

x = Flatten()(base_model.output)

output = Dense(len(Dataset_Labels), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

def data_generator(data, batch_size):
    while True:
        for i in range(0, len(data), batch_size):
            images, labels = [], []
            for j in range(i, min(i + batch_size, len(data))):
                images.append(data[j]['image'])
                labels.append(data[j]['labels'])
            yield np.stack(images), np.array(labels)

train_gen = data_generator(train_data, 32)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_gen, steps_per_epoch=len(train_data)//32, epochs=10)
````

````python
model.save('modelo.h5')
````

````python
# Modelo treinado
model = load_model('modelo.h5')

# Diretorio da imagem que você deseja prever
img_path = 'pikachu.png' 
img = image.load_img(img_path, target_size=(224, 224))

# Converta a imagem em um array numpy e expanda as dimensões para corresponder à entrada do modelo
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)

prediction = model.predict(img_tensor)

# O resultado é uma matriz com as probabilidades para cada classe, então pegamos o índice com a maior probabilidade
predicted_class = np.argmax(prediction)

print(f'A classe prevista é: {Dataset_Labels[predicted_class]}')
````
